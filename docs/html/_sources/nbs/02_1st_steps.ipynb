# 1st Steps

## Installation

To keep it short we just show code snippets. 
Assumptions:
- you have [conda or minicondainstalled](https://docs.conda.io/en/latest/miniconda.html).
- like to use [jupyter lab](https://jupyterlab.readthedocs.io/en/stable/getting_started/installation.html) as UI, otherwuse change code below accordingly

### create conda enviroment

use the following code in the conda prompt

```
conda create -n jupyterbook python=3.10
conda activate jupyterbook
conda install -c conda-forge jupyter-book
conda install pandas
conda install -c conda-forge matplotlib
conda install -c conda-forge jupyterlab
```

## 1st steps create book

- https://jupyterbook.org/en/stable/start/create.html

### create book template and test setup

Create a template book using the following code in the conda prompt
```
jupyter-book create mybook/
```

Test setup (don't move to directory mybook now)
```
jupyter-book build mybook/
```

check `index.html` in `_build/html/`

### housekeeping

because `jb build mybook/` created a template lets replace the defaults w.r.t. our needs

1. create folder for content files (notebooks, markdown,...), e.g. `Ç¹bs`
1. move `intro.md` to folder `nbs`
1. adjust `_toc.yml` such that header looks like
```
# Table of contents
# Learn more at https://jupyterbook.org/customize/toc.html

format: jb-book
root: nbs/intro
```
1. remove unessary template files

### Start page `intro.md`

This file will become the starting page of your book,, i.e. `index.html` in case of a web-book.

Hence add some content

### 1st content page

simply create a first content page, e.g. `01_content.ipynb`

#### Table of content `_toc.yml`

`_toc.yml` tells which files and in which order files become part of the book, e.g.:

```
# Table of contents
# Learn more at https://jupyterbook.org/customize/toc.html

format: jb-book
root: nbs/intro
chapters:
- file: nbs/01_content
```

or more information see [TOC-Doc](https://jupyterbook.org/en/stable/structure/toc.html).

#### `_config.yml`

define book settings, for more see [Configuration reference](https://jupyterbook.org/en/stable/customize/config.html)

1. adjust title
1. name author
1. (optional) remove full execution on build (`execute:` and `execute_notebooks: force`)
1. set repository location

1. add content
2. edit `_toc.yml`
